```python
# 숙제
# 1. normal(0, 1)에서 10개 샘플링
# 2. 10개의 평균값 구하기
# 3. 이 계산을 500 반복하고 결과를 리스트에 저장하기 -> 결과로 500개의 평균값이 얻어짐
# 4. 이 500개의 평균값에 대한 최솟값, 최댓값, 평균값 구하기
# 평균값 500개의 시각화하기: 히스토그램, 덴서티 플롯, 바이올린 플롯, 박스 플롯
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
def mean(data):
    mean = 0
    for n in data:
        mean += n
    mean /= len(data)
    return float(mean)

def repeat(num):
    mean_list = []
    for seed_num in range(num):
        np.random.seed(seed_num)  
        rand = np.random.normal(0, 1, nsamples) # 1. normal(0, 1)에서 10개 샘플링
        mean_list.append(float(mean(rand))) # 2. 10개의 평균값 구하기 + float로 변환 + mean_list에 넣기
    return mean_list

def cal_minimum(data_input):
    minimum = data_input[0]
    for n in data_input:
        if n < minimum:
            minimum = n
    return float(minimum)

def cal_maximum(data_input):
    maximum = data_input[0]
    for n in data_input:
        if n > maximum:
            maximum = n
    return float(maximum)

```

```python
niter = 500
nsamples = 10

result = repeat(niter) # 3. 계산을 500 반복하고 결과를 리스트에 저장하기 -> 결과로 500개의 평균값이 얻어짐, 시드 넘버 0~499까지.
print(len(result),"개 리스트의 평균")
print("최소:", cal_minimum(result))
print("최대:", cal_maximum(result))
print("평균:", mean(result)) # 4. 이 500개의 평균값에 대한 최솟값, 최댓값, 평균값 구하기
```

```python
# 평균값 500개의 시각화하기: 히스토그램, 덴서티 플롯

plt.hist(result, bins=30, density=True, edgecolor='black', alpha=0.5, color='blue')
sns.kdeplot(result, color='red', label='data1')
plt.title('Means of 500 samples')
plt.show()
```

```python
# 바이올린 플롯
plt.violinplot(result, showmeans=True, showextrema=True)
plt.grid(True)
plt.show()
```

```python
# 박스 플롯
plt.boxplot(result)
plt.show()
```
