```python
import random
import matplotlib.pyplot as plt

numbers = []
for i in range(12):
    r = random.randint(-50, 50)
    numbers.append(r)

print(numbers)

plt.plot(numbers,marker='o')
plt.show()
```

```python
# 평균 값 구하기
총합 = numbers[0] + numbers[1] + numbers[2] + numbers[3] + numbers[4] + numbers[5] + numbers[6] + numbers[7] + numbers[8] + numbers[9] + numbers[10] + numbers[11] 
평균값 = 총합/len(numbers)
print("총합:", 총합, "평균값:", 평균값)
```

```python
total = 0
for num in numbers:
    total += num
average = total / len(numbers)
print("총합:", total, "평균값:", average)
```

```python
def mean(nums):
    return sum(nums) / len(nums)

m = mean(numbers)
print("평균값:", m)
```

정규분포 = normal distribution = Gaussian distribution

```python
import matplotlib.pyplot as plt
standard_normal_samples = np.random.randn(1000)
plt.plot(standard_normal_samples,marker='o',markersize=3)
plt.show()
```

```python
plt.hist(standard_normal_samples, bins=50, edgecolor='black')
```

```python
print("평균:",standard_normal_samples.mean())
print("표편:",standard_normal_samples.std())
```

```python
samples_10000 = np.random.normal(loc=0,scale=1,size=10000)
plt.hist(samples_10000, bins=50, edgecolor='black')
plt.show()
```

```python
samples_10000_scale2 = np.random.normal(loc=0,scale=2,size=10000)
plt.hist(samples_10000_scale2, bins=50, edgecolor='black')
plt.show()
```

```python
samples_10000_loc4 = np.random.normal(loc=4,scale=1,size=10000)
plt.hist(samples_10000_loc4, bins=50, edgecolor='black')
plt.show()
```

```python
plt.figure(figsize=(10,6))
plt.plot(samples_10000, label='samples_10000', alpha=0.5)
plt.plot(samples_10000_scale2, label='samples_10000_scale2', alpha=0.5)
plt.plot(samples_10000_loc4, label='samples_10000_loc4', alpha=0.5)
plt.legend()
plt.title('통합 그래프')
plt.xlabel('Index')
plt.ylabel('Value')
plt.show()
```
