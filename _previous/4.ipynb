```python
data3 = np.random.normal(loc=5, scale=2, size=249)
data4 = np.random.laplace(loc=0, scale=1, size=10000)
data5 = np.random.exponential(scale=1, size=10000)
```

```python
datalist = [data1, data2, data3, data4, data5]
plt.violinplot(datalist, showmeans=True, showextrema=True)
plt.grid(True)
plt.show()
```

```python
plt.boxplot(datalist, showmeans=True)

# Add IQR and outliers annotation for each dataset
for i, data in enumerate(datalist, start=1):
    q1 = np.percentile(data, 25)
    q3 = np.percentile(data, 75)
    iqr = q3 - q1
    # Outliers: outside 1.5*IQR from Q1/Q3
    lower_bound = q1 - 1.5 * iqr
    upper_bound = q3 + 1.5 * iqr
    outliers = data[(data < lower_bound) | (data > upper_bound)]
    plt.text(i, q3, f'IQR={iqr:.2f}', ha='center', va='bottom', fontsize=8, color='blue')
    if len(outliers) > 0:
        plt.text(i, q3 + 0.2 * iqr, f'Outliers={len(outliers)}', ha='center', va='bottom', fontsize=8, color='red')
plt.grid(True)
plt.show()
```

```python
sa = np.sort(s)
plt.plot(s)
plt.plot(sa)
plt.show()
```

```python
plt.violinplot(s, showmeans=True, showextrema=True)
q1 = np.percentile(s, 25)
q3 = np.percentile(s, 75)
iqr = q3 - q1
plt.plot([1, 1], [q1, q3], color='red', linestyle='--', label='IQR')
plt.axhspan(q1, q3, color='lightblue', alpha=0.5, label='IQR')
plt.axhline(q3, color='purple', linestyle='--', label='Q3')
plt.text(1.05, q3, f'IQR={iqr:.2f}', va='bottom', fontsize=10, color='blue')
plt.legend()
plt.show()
```
